```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
wisc = pd.read_csv('c:/data/wisc_bc_data.csv')
wisc.head()
```

```python
wisc.diagnosis.value_counts()
```

```python
wisc.columns
```

```python
plt.scatter(range(357),'radius_mean',data=wisc[wisc.diagnosis=='B'],c='green',s=5,label='Benign')
plt.scatter(range(212),'radius_mean',data=wisc[wisc.diagnosis=='M'],c='red',s=5,label='Malignant')
plt.legend()
plt.xlabel('Index')
plt.ylabel('Radius  Mean')
plt.show()
```

```python
plt.scatter(range(357),'area_mean',data=wisc[wisc.diagnosis=='B'],c='green',s=5,label='Benign')
plt.scatter(range(212),'area_mean',data=wisc[wisc.diagnosis=='M'],c='red',s=5,label='Malignant')
plt.legend()
plt.xlabel('Index')
plt.ylabel('area  Mean')
plt.show()
```

```python
columns = ['radius_mean', 'texture_mean','smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']
for c in columns:
    plt.scatter(range(357),c,data=wisc[wisc.diagnosis=='B'],c='green',s=5,label='Benign')
    plt.scatter(range(212),c,data=wisc[wisc.diagnosis=='M'],c='red',s=5,label='Malignant')
    plt.legend()
    plt.xlabel('Index')
    plt.ylabel(c)
    plt.show()
           
```

```python
columns=['radius_worst', 'texture_worst','smoothness_worst','compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']
for c in columns:
    plt.scatter(range(357),c,data=wisc[wisc.diagnosis=='B'],c='green',s=5,label='Benign')
    plt.scatter(range(212),c,data=wisc[wisc.diagnosis=='M'],c='red',s=5,label='Malignant')
    plt.legend()
    plt.xlabel('Index')
    plt.ylabel(c)
    plt.show()
```
